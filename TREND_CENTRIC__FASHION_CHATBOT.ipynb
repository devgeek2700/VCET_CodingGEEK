```python
# !pip install PyMuPDF==1.18.19

import fitz  # Import the PyMuPDF library, aliased as fitz

# Function to extract text from PDFs
def extract_texts_from_pdfs(pdf_paths):
    all_texts = []
    for pdf_path in pdf_paths:
        doc = fitz.open(pdf_path)
        text = ''
        for page in doc:
            text += page.get_text()
        all_texts.append(text)
    return all_texts
```

```python
# Imports for data ingestion and processing
import os
import requests
from bs4 import BeautifulSoup

# Function to scrape websites recursively
def scrape_websites_recursive(urls, depth=1):
    all_texts = []
    seen_urls = set()

    def scrape(url, current_depth):
        if url in seen_urls or current_depth > depth:
            return
        seen_urls.add(url)

        # Check if the URL has a valid scheme
        if not url.startswith('http'):
            return  # Skip if it's not a valid web URL

        response = requests.get(url)
        if response.status_code == 200:
            soup = BeautifulSoup(response.content, 'html.parser')
            texts = ' '.join([p.text for p in soup.find_all('p')])
            all_texts.append(texts)
            # Extract and follow links
            links = [a['href'] for a in soup.find_all('a', href=True)]
            for link in links:
                # Handle relative URLs
                if link.startswith('/'):
                    link = requests.compat.urljoin(url, link)
                scrape(link, current_depth + 1)
        else:
            print(f"Failed to scrape {url}")

    for url in urls:
        scrape(url, 0)

    return all_texts
```

```python
import math
# Function to chunk documents into smaller segments
def chunk_documents(documents, chunk_size=500):
    chunked_documents = []
    for doc in documents:
        num_chunks = math.ceil(len(doc) / chunk_size)
        chunks = [doc[i*chunk_size:(i+1)*chunk_size] for i in range(num_chunks)]
        chunked_documents.extend(chunks)
    return chunked_documents

# Chunking documents
chunked_documents = chunk_documents(documents)
```

```python
# Imports for embedding and indexing
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
```

```python
# Function to create embeddings using Sentence Transformers
def create_embeddings(documents):
    model = SentenceTransformer('all-MiniLM-L6-v2')
    embeddings = model.encode(documents)
    return embeddings, model

# Function to create FAISS index
def create_faiss_index(embeddings):
    dimension = embeddings.shape[1]
    index = faiss.IndexFlatL2(dimension)
    index.add(embeddings)
    return index
```

```python
# Creating embeddings and FAISS index
embeddings, _ = create_embeddings(documents)
index = create_faiss_index(embeddings)
```

```python
# Save the embeddings and index to disk
faiss.write_index(index, "index.faiss")

# Load the FAISS index
index = faiss.read_index("index.faiss")
```

#RagChatbot

```python
# Function to generate response using RAG pipeline
def rag_chatbot(query):
    inputs_dict = tokenizer(
        query,
        return_tensors="pt",
        padding=True,
        truncation=True,
        max_length=512,
    )

    # Retrieve relevant documents
    retriever_outputs = retriever(**inputs_dict)
    doc_ids = retriever_outputs["context_input_ids"]

    # Generate response
    generator_inputs = {
        "input_ids": doc_ids,
        "attention_mask": inputs_dict["attention_mask"]
    }
    generator_outputs = generator.generate(
        **generator_inputs,
        max_length=150,
    )

    return tokenizer.decode(generator_outputs[0], skip_special_tokens=True)
```

```python
# Gradio interface code
def gradio_chatbot_interface():
    def chat_with_bot(user_input):
        return rag_chatbot(user_input)

    iface = gr.Interface(
        fn=chat_with_bot,
        inputs="text",
        outputs="text",
        title="RAG Chatbot",
        description="Welcome to the RAG Chatbot. Type your query and press Enter to chat."
    )

    iface.launch(share=True)

# Run the Gradio app
gradio_chatbot_interface()
```
